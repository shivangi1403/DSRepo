# Task:

1.Consolidate these 50 datasets into a single dataset in a meaningful way.

2.Append Province onto the consolidated dataset.

```python
import glob
import pandas as pd

path = '/content'
csv_files = glob.glob(path + "/*.csv")
file_to_remove = '/content/id_prov.csv'
csv_files.remove(file_to_remove)
```

```python
# Read each CSV file into DataFrame
# This creates a list of dataframes
df_list = (pd.read_csv(file) for file in csv_files)

# Consolidating the 50 datasets into a single dataset in a meaningful way.
big_df = pd.concat(df_list, ignore_index=True)
```

```python
# Total number of records in the DataFrame
num_records = big_df.shape[0]
print(num_records)
```

```python
# Verifying count of NAN in create_dt column
nan_count = big_df['create_dt'].isna().sum()
print(nan_count)
```

```python
big_df = big_df.drop('create_dt', axis=1)
```

```python
big_df.head(100)
```

```python
# Verifying count of NAN for all the other columns
nan_id1 = big_df['id1'].isna().sum()
print(nan_id1)
nan_id2 = big_df['id2'].isna().sum()
print(nan_id2)
nan_part_dt = big_df['part_dt'].isna().sum()
print(nan_part_dt)
```

```python
# Chceking data type of columns to verify any data type needs ro be changd.
print("Column datatypes: ")
print(big_df.dtypes)
```

```python
big_df = big_df.dropna(subset=['part_dt'])
```

```python
# Total number of records after cleaning data in the DataFrame

num_records = big_df.shape[0]
print(num_records)
```

```python
# Append Province onto the consolidated dataset.
df_prov = pd.read_csv('/content/id_prov.csv')
big_df = big_df.merge(df_prov, on=['id1','id2'])
```
